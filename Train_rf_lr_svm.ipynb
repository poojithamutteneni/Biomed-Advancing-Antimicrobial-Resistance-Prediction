```python
# Load data
data = pd.read_csv('data.csv', index_col='prename')
pheno = pd.read_csv('pheno.csv', index_col='prename')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split, cross_val_predict, StratifiedKFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, roc_curve, auc, matthews_corrcoef, confusion_matrix


# Standardize features
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

# Apply PCA
pca = PCA(n_components=0.95)  # Retain 95% of variance
data_pca = pca.fit_transform(data_scaled)

# Initialize models
rf = RandomForestClassifier(n_estimators=500, max_depth=10, min_samples_split=5, min_samples_leaf=2)
lr = LogisticRegression(max_iter=1000, C=0.1, solver='lbfgs')
svm = SVC(kernel='linear', C=0.01)

labels = ['CIP', 'CTX', 'CTZ', 'GEN']
for label in labels:
    print(f"\nClassification for {label}:")
    # Label encoding for the current label
    y = pheno[label]

    # Perform 5-fold cross-validation and prediction
    rf_pred = cross_val_predict(rf, data_pca, y, cv=5)
    lr_pred = cross_val_predict(lr, data_pca, y, cv=5)
    svm_pred = cross_val_predict(svm, data_pca, y, cv=5)

    # Calculate accuracy scores
    rf_accuracy = accuracy_score(y, rf_pred)
    lr_accuracy = accuracy_score(y, lr_pred)
    svm_accuracy = accuracy_score(y, svm_pred)

    # Print accuracy scores
    print("Random Forest Accuracy:", rf_accuracy)
    print("Logistic Regression Accuracy:", lr_accuracy)
    print("Support Vector Machine Accuracy:", svm_accuracy)

    # Generate classification reports
    rf_report = classification_report(y, rf_pred)
    lr_report = classification_report(y, lr_pred)
    svm_report = classification_report(y, svm_pred)

    # Print classification reports
    print("\nRandom Forest Classification Report:")
    print(rf_report)
    print("\nLogistic Regression Classification Report:")
    print(lr_report)
    print("\nSupport Vector Machine Classification Report:")
    print(svm_report)

    # Calculate MCC for each fold of cross-validation
    rf_mcc = [matthews_corrcoef(y[test], rf_pred[test]) for _, test in StratifiedKFold(n_splits=5).split(data_pca, y)]
    lr_mcc = [matthews_corrcoef(y[test], lr_pred[test]) for _, test in StratifiedKFold(n_splits=5).split(data_pca, y)]
    svm_mcc = [matthews_corrcoef(y[test], svm_pred[test]) for _, test in StratifiedKFold(n_splits=5).split(data_pca, y)]

    # Print MCC for each fold
    print("\nRandom Forest MCC:", rf_mcc)
    print("Logistic Regression MCC:", lr_mcc)
    print("Support Vector Machine MCC:", svm_mcc)

    # Generate confusion matrices for each fold of cross-validation
    rf_cm = [confusion_matrix(y[test], rf_pred[test]) for _, test in StratifiedKFold(n_splits=5).split(data_pca, y)]
    lr_cm = [confusion_matrix(y[test], lr_pred[test]) for _, test in StratifiedKFold(n_splits=5).split(data_pca, y)]
    svm_cm = [confusion_matrix(y[test], svm_pred[test]) for _, test in StratifiedKFold(n_splits=5).split(data_pca, y)]

    # Print confusion matrices for each fold
    print("\nRandom Forest Confusion Matrix:", rf_cm)
    print("Logistic Regression Confusion Matrix:", lr_cm)
    print("Support Vector Machine Confusion Matrix:", svm_cm)

    # Calculate ROC curve and AUC for each classifier
    plt.figure(figsize=(8, 6))
    for name, pred in [("Random Forest", rf_pred), ("Logistic Regression", lr_pred), ("Support Vector Machine", svm_pred)]:
        fpr, tpr, _ = roc_curve(y, pred)
        roc_auc = auc(fpr, tpr)
        plt.plot(fpr, tpr, label=f"{name} (AUC = {roc_auc:.2f})")
    plt.plot([0, 1], [0, 1], linestyle='--', color='gray')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'ROC Curve for {label}')
    plt.legend()
    plt.show()

    # Plot MCC values
    plt.figure(figsize=(8, 6))
    plt.plot(range(1, 6), rf_mcc, label='Random Forest')
    plt.plot(range(1, 6), lr_mcc, label='Logistic Regression')
    plt.plot(range(1, 6), svm_mcc, label='Support Vector Machine')
    plt.xlabel('Fold')
    plt.ylabel('MCC Value')
    plt.title(f'MCC Values for {label}')
    plt.legend()
    plt.show()

    # Plot confusion matrices
    plt.figure(figsize=(16, 6))
    for i, cm in enumerate([rf_cm, lr_cm, svm_cm]):
        plt.subplot(1, 3, i + 1)
        avg_cm = np.mean(cm, axis=0)
        plt.imshow(avg_cm, interpolation='nearest', cmap=plt.cm.Blues)
        plt.colorbar()
        plt.title(f'Average Confusion Matrix - {labels[i]}')
        tick_marks = np.arange(len(labels))
        plt.xticks(tick_marks, labels, rotation=45)
        plt.yticks(tick_marks, labels)
        plt.tight_layout()
        plt.ylabel('True label')
        plt.xlabel('Predicted label')
    plt.show()
```
